```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
numeric_columns = df_market_data.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_scaled_market_data = pd.DataFrame(scaled_data, columns=numeric_columns, index=df_market_data.index)



# Copy the crypto names from the original DataFrame
df_scaled_market_data['coin_id'] = df_market_data.index


# Set the coin_id column as index

df_scaled_market_data.set_index('coin_id', inplace=True)


# Display the scaled DataFrame

df_scaled_market_data.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

# Loop through k-values from 1 to 11
for k in range(1, 12):
    # Create a KMeans model with the current number of clusters (k)
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Fit the model to the scaled data
    kmeans.fit(df_scaled_market_data)
    
    # Append the inertia value to the list
    inertia.append(kmeans.inertia_)

# Display the inertia values
print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": list(range(1, 12)),  # List of k-values
    "inertia": inertia       # List of inertia values
}


# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

print(df_elbow)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.


# Plot the Elbow curve with hvPlot
elbow_plot = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve to Identify Optimal k", 
    xlabel="Number of Clusters (k)", 
    ylabel="Inertia", 
    width=800, 
    height=400
)

# Show the plot
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 

The best k value is 4.

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans.fit(df_scaled_market_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame

predicted_clusters = kmeans.predict(df_scaled_market_data)


# Print the resulting array of cluster values.
print("Predicted Cluster Labels:\n", predicted_clusters)
```

```python
# Create a copy of the scaled DataFrame

df_scaled_market_data_copy = df_scaled_market_data.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters

df_scaled_market_data_copy['cluster'] = predicted_clusters

# Display the copy of the scaled DataFrame
df_scaled_market_data_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

# Create a scatter plot using hvPlot
scatter_plot = df_scaled_market_data_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    c="cluster",  # Color points based on cluster labels
    colormap='Category10',  # Choose a color map
    hover_cols=['crypto_name'],  # Show cryptocurrency name in hover
    title="Cryptocurrency Price Change vs Cluster",
    xlabel="24h Price Change (%)", 
    ylabel="7d Price Change (%)",
    width=800, 
    height=400
)

# Display the plot
scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_scaled_market_data)
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])


# View the scaled PCA data
df_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.

explained_variance = pca.explained_variance_ratio_

# Display the explained variance for each principal component
print("Explained Variance Ratio for each principal component:\n", explained_variance)

```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Total Explained Variance: 0.8950316570309841

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])


# Copy the crypto names from the original scaled DataFrame

df_pca['coin_id'] = df_market_data.index


# Set the coin_id column as index

df_pca.set_index('coin_id', inplace=True)  # Set 'coin_id' as the index


# Display the scaled PCA DataFrame
df_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values

inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list


# Loop through the k-values from 1 to 11
for k in range(1, 12):
    # Initialize the KMeans model with the current value of k
    kmeans = KMeans(n_clusters=k, random_state=42)  # Using random_state for reproducibility
    
    # Fit the model to the PCA-transformed data (df_market_data_pca)
    kmeans.fit(df_pca)  # Assuming df_pca is the PCA-transformed DataFrame
    
    # Append the inertia value to the list
    inertia_values.append(kmeans.inertia_)

# Display the inertia values for each k
print(inertia_values)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_data = {
    "k": list(range(1, 12)),  # List of k-values
    "inertia": inertia_values      # List of inertia values
}

# Create a DataFrame with the data to plot the Elbow curve

elbow_df = pd.DataFrame(elbow_data)

```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.

# Plot the Elbow curve with hvPlot
elbow_plot = elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve for Optimal k", 
    xlabel="Number of Clusters (k)", 
    ylabel="Inertia", 
    width=800, 
    height=400
)

# Show the plot
elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It does not differ.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
optimal_k = 4  
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
kmeans.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
cluster_predictions = kmeans.predict(df_pca)


# Print the resulting array of cluster values.
print("Cluster predictions (labels):\n", cluster_predictions)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_copy = df_pca.copy()


# Add a new column to the copy of the PCA DataFrame with the predicted clusters

df_pca_copy['cluster'] = cluster_predictions

# Display the copy of the scaled PCA DataFrame
df_pca_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

df_pca_copy['crypto_name'] = df_market_data.index

# Create a scatter plot using hvPlot
df_pca_copy.hvplot.scatter(
    x='PC1', 
    y='PC2', 
    c='cluster',  # Color by K-Means cluster labels
    cmap='Set1',  # Choose a color map for the clusters
    title='PCA Clusters of Cryptocurrencies',  # Title of the plot
    xlabel='PC1',  # Label for the x-axis
    ylabel='PC2',  # Label for the y-axis
    hover_cols=['crypto_name'],  # Show crypto names on hover
    size=8,  # Set the size of the points
    legend='top_right'  # Position the legend in the top-right corner
)
```
